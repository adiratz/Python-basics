# Making a grid and adding multiple charts( columns and rows of charts)

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.FacetGrid(tips, col="sex")
```

```python
g= sns.FacetGrid(tips, col="sex")
g.map(sns.distplot, "total_bill")
```

```python
g= sns.FacetGrid(tips, row="sex")
g.map(sns.distplot, "total_bill")
```

```python
g= sns.FacetGrid(tips, row="sex", col="smoker")
g.map(sns.distplot, "total_bill")
```
