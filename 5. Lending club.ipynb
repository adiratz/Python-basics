```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("LoanStats2018Q1.csv", skiprows=1)
```

```python
df.shape
```

```python
## average funded amount for each term
```

```python
df.term.value_counts() ##how many terms
```

```python
sns.catplot("term", data=df, kind="count")
```

```python
df_grouped= df.groupby("term")
```

```python
df_grouped.funded_amnt.mean()
```

```python
sns.catplot("term", "funded_amnt", data=df, kind= "box" )
```

```python
sns.catplot("grade", "funded_amnt", data=df, kind= "box", hue="term")
```

```python
sns.catplot("grade", "funded_amnt", data=df, kind= "box", hue="term", order=["A", "B", "C", "D", "E", "F", "G"])
```
